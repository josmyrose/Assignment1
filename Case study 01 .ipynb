```python
#Read the dataset to the python environment.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
car_df=pd.read_csv('cars_data.csv')
```

```python
#Get the details of  car data set
car_df.head()
```

```python
car_df.info()
```

```python

# Check for the null values present in the dataset.
car_df.isnull().sum()
```

```python
#Plot a bar graph of male vs female buyers participated in the sales.
plt.hist(car_df['Buyer Gender'],bins=4)
```

```python
car_df['Buyer Gender'].value_counts()
```

```python
#We can deduce from the graph above that there isn't much of a difference in buyer gender.
```

```python
#Find the top 5 cars based on their sales price.
car_df.sort_values(by=['Sale Price'],na_position='first').head(5)
#df.sort_values(by=['Population'], na_position='first')
```

```python
car_df.drop_duplicates(subset ="Sale Price",keep = False, inplace = True) 
```

```python
car_df.drop_duplicates()
```

```python
#Find the least 5 cars based on their Resell price
car_df.sort_values(by=['Resell Price'],na_position='first').head(5)
```
